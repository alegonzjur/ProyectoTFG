### Gráfica que muestre el crecimiento de muertes por una enfermedad en España.

```python
españa = data.query('Entity == "Spain"')
españa
```

```python
mental_fatality = input("Introduzca un nombre de una enfermedad mental:")
fisical_fatality = input("Introduzca el nombre de una lesión fisica:")
x = españa[mental_fatality]
z = españa[fisical_fatality]
y = españa['Year']
fig, ax = plt.subplots(figsize = (8,7))
ax.plot(y,x, color='Blue')
ax.plot(y,z, color='Orange')
```
